```python
import requests
import pandas as pd
import time

foods = [
    "chicken", "beef", "lamb", "turkey", "duck", "fish", "salmon", "tuna", "shrimp",
    "rice", "pasta", "bread", "oats", "potato", "sweet potato", "quinoa", "corn",
    "apple", "banana", "orange", "grape", "strawberry", "mango", "pineapple", "watermelon",
    "tomato", "carrot", "cucumber", "lettuce", "spinach", "broccoli", "onion", "garlic",
    "milk", "cheese", "yogurt", "butter", "cream",
    "pizza", "burger", "sandwich", "sushi", "shawarma", "kebab", "fried chicken",
    "lasagna", "noodles", "falafel", "hummus", "burrito", "taco",
    "cookies", "cake", "chocolate", "chips", "popcorn",
    "coffee", "tea", "juice", "smoothie"
]

def get_meals(food):
    url = f"https://www.themealdb.com/api/json/v1/1/search.php?s={food}"
    res = requests.get(url).json()
    return res.get("meals", [])

def extract_ingredients(meal):
    ingredients = []

    for i in range(1, 21):
        ing = meal.get(f"strIngredient{i}")
        if ing and ing.strip():
            ingredients.append(ing.strip())

    return ingredients


dataset = []

for food in foods:
    meals = get_meals(food)

    if not meals:
        continue

    for meal in meals:
        ingredients = extract_ingredients(meal)

        dataset.append({
            "food_query": food,
            "meal_name": meal["strMeal"],
            "ingredients": ", ".join(ingredients)
        })

    print(f"{food} done → total: {len(dataset)}")
    time.sleep(0.5)


df = pd.DataFrame(dataset)

df.drop_duplicates(inplace=True)

df.to_csv("ingredients_full_dataset.csv", index=False)

print("FINAL SHAPE:", df.shape)
print(df.head())
```

```python
import requests
import pandas as pd
import time
import random

dataset = []

# -------------------------
# extract ingredients
# -------------------------
def extract_ingredients(meal):
    ingredients = []

    for i in range(1, 21):
        ing = meal.get(f"strIngredient{i}")
        if ing and ing.strip():
            ingredients.append(ing.strip().lower())

    return ingredients


# -------------------------
# 1. search by letters
# -------------------------
def get_by_letter(letter):
    url = f"https://www.themealdb.com/api/json/v1/1/search.php?f={letter}"
    res = requests.get(url).json()
    return res.get("meals", [])


# -------------------------
# 2. random meals
# -------------------------
def get_random():
    url = "https://www.themealdb.com/api/json/v1/1/random.php"
    res = requests.get(url).json()
    return res.get("meals", [])


# -------------------------
# STEP 1: letters
# -------------------------
for letter in "abcdefghijklmnopqrstuvwxyz":
    meals = get_by_letter(letter)

    if meals:
        for meal in meals:
            dataset.append({
                "name": meal["strMeal"],
                "ingredients": ", ".join(extract_ingredients(meal))
            })

    print(f"Letter {letter} done → {len(dataset)}")
    time.sleep(0.3)


# -------------------------
# STEP 2: random (BIG BOOST)
# -------------------------
for i in range(15000):
    meals = get_random()

    if meals:
        meal = meals[0]

        dataset.append({
            "name": meal["strMeal"],
            "ingredients": ", ".join(extract_ingredients(meal))
        })

    if i % 500 == 0:
        print(f"Random {i} → total {len(dataset)}")

    time.sleep(0.2)


# -------------------------
# CLEANING
# -------------------------
df = pd.DataFrame(dataset)

df.drop_duplicates(inplace=True)

df = df[df["ingredients"].str.len() > 5]

print("FINAL SIZE:", df.shape)

df.to_csv("ingredients_15k_dataset.csv", index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {
    "User-Agent": "Mozilla/5.0"
}

category_urls = {
    "pizza": "https://tasty.co/search?q=pizza&sort=popular",
    "burger": "https://tasty.co/search?q=burger&sort=popular",
    "chicken": "https://tasty.co/search?q=chicken&sort=popular",
    "rice": "https://tasty.co/search?q=rice&sort=popular",
    "apple": "https://tasty.co/search?q=apple&sort=popular",
    "coffee": "https://tasty.co/search?q=coffee&sort=popular",
    "sushi": "https://tasty.co/search?q=sushi&sort=popular",
}
category_urls.update({
    # chocolate
    "chocolate": "https://tasty.co/search?q=chocolate&sort=popular",
    "cocoa": "https://tasty.co/search?q=cocoa&sort=popular",
    "chocolate cake": "https://tasty.co/search?q=chocolate+cake&sort=popular",

    # beans
    "beans": "https://tasty.co/search?q=beans&sort=popular",
    "black beans": "https://tasty.co/search?q=black+beans&sort=popular",
    "kidney beans": "https://tasty.co/search?q=kidney+beans&sort=popular",
    "lentils": "https://tasty.co/search?q=lentils&sort=popular",

    # extras (مهمة للdataset بتاعك)
    "vanilla": "https://tasty.co/search?q=vanilla&sort=popular",
    "honey": "https://tasty.co/search?q=honey&sort=popular",
    "cinnamon": "https://tasty.co/search?q=cinnamon&sort=popular"
})

all_data = []

def get_soup(url):
    try:
        res = requests.get(url, headers=headers, timeout=10)
        if res.status_code != 200:
            return None
        return BeautifulSoup(res.text, "html.parser")
    except:
        return None


for category, url in category_urls.items():
    print(f"\nScraping: {category}")

    soup = get_soup(url)
    if not soup:
        continue

    links = soup.find_all("a", href=True)

    recipes = set()

    for link in links:
        href = link["href"]

        if "/recipe/" in href:
            name = link.get_text(strip=True)

            if name:
                recipes.add((name, "https://tasty.co" + href))

    print(f"Found {len(recipes)} recipes")

    for name, link in list(recipes)[:30]:  # limit for stability
        soup2 = get_soup(link)
        if not soup2:
            continue

        ingredients = soup2.find_all("li")

        ing_list = []
        for i in ingredients:
            text = i.get_text(strip=True)
            if any(word in text.lower() for word in ["cup", "tsp", "tbsp", "salt", "oil", "flour", "egg", "milk"]):
                ing_list.append(text)

        if ing_list:
            all_data.append([name, category, ", ".join(ing_list)])

        time.sleep(0.5)

df = pd.DataFrame(all_data, columns=["Recipe Name", "Category", "Ingredients"])
df.drop_duplicates(inplace=True)

df.to_csv("tasty_FINAL2_REAL.csv", index=False)

print(f"\nDONE ✔ Total recipes: {len(df)}")
```
